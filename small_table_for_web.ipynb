```python
vdname_table=['NVQKWL40','SVQKWL40','NVQ6WK40','SVQ6WK40','NVQKWL00','SVQKWL00','UVSSWC60']

import time
tStart = time.time()

db_connection = sql.connect(host='localhost', database='for1121', user='root', password='apple')
frames = []
for ele in vdname_table:
    frames.append(pd.read_sql('SELECT * FROM {r}'.format(r=ele), con=db_connection))

result = pd.concat(frames, ignore_index=True)

tEnd = time.time()
print "It cost %f sec" % (tEnd - tStart)
```

```python
#把bytesarray 變成utf-8
def decode_to_utf8(df):
    df.EQIPnumber=df.EQIPnumber.str.decode("utf-8")
    df.location=df.location.str.decode("utf-8")
    df.direct=df.direct.str.decode("utf-8")
    df.YMD=df.YMD.str.decode("utf-8")
    df.hour=df.hour.str.decode("utf-8")
    return df

df_all=decode_to_utf8(result) 
```

```python
def define_road(df):
    import time
    tStart = time.time()
    
    df.ix[:,'jb_result'] = None
    jb.load_userdict("0315road_dict.txt")
    for locatnum in range(len(df.ix[:,'location'])):
        jb_cut = jb.cut(df.ix[locatnum,'location'])
        temp= []
        for i in jb_cut:
            if (('路'in i) or ('街' in i) or ('大道' in i)or('橋' in i))&(len(i)>2):
                temp.append(i)
        temp=";".join(temp)
        df.set_value(locatnum,'jb_result',temp)
        
    tEnd = time.time()
    print "It cost %f sec" % (tEnd - tStart)
```

```python
define_road(df_all)
```
